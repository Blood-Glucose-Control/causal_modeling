```python
import pandas as pd
pd.set_option('display.max_columns', None)

df_gl = pd.read_csv('data/500030_5th-7th.csv')
df_daily_readiness = pd.read_csv('data/fitness/DailyReadiness/Daily Readiness Score - 2024-04-01.csv')
df_hrv_details = pd.read_csv('data/fitness/HeartRateVariability/HRVDetails/Heart Rate Variability Details - 2024-04-03.csv')
df_hrv_summ = pd.read_csv('data/fitness/HeartRateVariability/HRVSummary/Daily Heart Rate Variability Summary - 2024-04-.csv')
df_hrv_resp = pd.read_csv('data/fitness/HeartRateVariability/RespiratoryRateSummary/Daily Respiratory Rate Summary - 2024-04-01.csv')
df_sleep_score = pd.read_csv('data/fitness/SleepScore/sleep_score.csv')
df_sleep_prof = pd.read_csv('data/fitness/SleepScore/Sleep Profile.csv')
df_daily_spo2 = pd.read_csv('data/fitness/SPO2/Daily SpO2 - 2023-11-15-2024-02-23.csv')
df_min_spo2 = pd.read_csv('data/fitness/SPO2/Minute SpO2 - 2024-01-01.csv')
df_stress_score = pd.read_csv('data/fitness/StressScore/Stress Score.csv')
df_tempor_comp = pd.read_csv('data/fitness/Temperature/Computed Temperature - 2024-01-01.csv')
df_tempor_device = pd.read_csv('data/fitness/Temperature/Device Temperature - 2024-01-02.csv')

df_output = pd.read_csv('output/merged_health_data.csv')
```

## Timezone Awareness

Timezone Aware
- BGL (df_gl): `2024-05-01 00:00:27.029000-05:00`
- SPO2 (df_daily_spo2): `2023-11-16T00:00:00Z`
- Sleep Score (df_sleep_score): `2024-11-24T10:38:30Z`

Not Timezone Aware
- HRV Details (df_hrv_details): `2024-04-03T02:50:00`
- HRV Summary (df_hrv_summ): `2024-04-01T00:00:00`
- HRV Respiratory (df_hrv_resp): `2024-04-01T00:00:00`
- Temperature Computed (df_tempor_comp): `2024-01-01T05:03`
- Temperature Device (df_tempor_device): `2024-01-02T00:00`

Date Only
- Sleep Profile (df_sleep_prof): `2023-12-30`
- Daily Readiness (df_daily_readiness): `2024-05-01`
