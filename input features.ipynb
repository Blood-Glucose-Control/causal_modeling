## Input Data Dictionary
---

### Blood Glucose Dataset
- `bgl_date`: Timestamp with timezone (UTC-05:00)
- `bgl_sender_id`: Unique identifier for the sender
- `bgl_value`: Blood glucose level measurement
- `bgl_date_millis`: Timestamp in milliseconds
- `bgl_text`: Description or notes about the reading (what meal was eaten)
- `bgl_template`: Template type for the message
- `bgl_msg_type`: Type of message (e.g., DOSE_INSULIN, ANNOUNCE_MEAL)
- `bgl_affects_fob`: Boolean indicating if affects future carbs
- `bgl_affects_iob`: Boolean indicating if affects insulin on board
- `bgl_dose_units`: Units of insulin dosed
- `bgl_food_g`: Grams of carbohydrates consumed
- `bgl_food_glycemic_index`: Glycemic index of consumed food
- `bgl_dose_automatic`: Boolean indicating if dose was automatic
- `bgl_trend`: Blood glucose trend (e.g., FLAT, UP, DOWN)

### Daily Readiness Dataset
- `readiness_timestamp`: UTC timestamp
- `readiness_score_value`: Daily readiness score (0-100)
- `readiness_state`: Current readiness state

### Sleep Score Dataset
- `sleep_timestamp`: UTC timestamp
- `sleep_overall_score`: Overall sleep score (0-100)
- `sleep_composition_score`: Sleep composition score
- `sleep_revitalization_score`: Sleep revitalization score
- `sleep_duration_score`: Sleep duration score
- `sleep_deep_sleep_minutes`: Minutes spent in deep sleep
- `sleep_resting_heart_rate`: Resting heart rate during sleep
- `sleep_restlessness`: Measure of sleep restlessness

### Sleep Profile Dataset
- `sleep_profile_date`: Creation date
- `sleep_type`: Type of sleep
- `deep_sleep`: Deep sleep percentage
- `rem_sleep`: REM sleep percentage
- `sleep_duration`: Total sleep duration (hours)
- `sleep_start_time`: Time sleep started
- `schedule_variability`: Variability in sleep schedule (hours)
- `restorative_sleep`: Restorative sleep score (0-100)
- `time_before_sound_sleep`: Minutes before sound sleep
- `sleep_stability`: Wake counts per hour
- `nights_with_long_awakenings`: Percentage of nights with long awakenings
- `days_with_naps`: Percentage of days with naps

### Temperature Dataset
- `temp_timestamp`: Recording timestamp
- `temp_value`: Temperature measurement
- `temp_nightly`: Nightly temperature average
- `temp_samples`: Number of temperature samples
- `temp_baseline_deviation`: Deviation from baseline temperature

### SPO2 Dataset
- `spo2_timestamp`: UTC timestamp
- `spo2_average`: Average SpO2 value
- `spo2_lower_bound`: Lower bound of SpO2 measurement
- `spo2_upper_bound`: Upper bound of SpO2 measurement

### Stress Score Dataset
- `stress_date`: UTC timestamp
- `stress_score`: Overall stress score
- `stress_sleep_points`: Sleep component of stress score
- `stress_responsiveness`: Responsiveness score
- `stress_exertion`: Exertion score
- `stress_status`: Status of stress measurement
- `stress_calculation_failed`: Boolean indicating calculation failure

### HRV Dataset
- `hrv_timestamp`: UTC timestamp
- `hrv_rmssd`: Root mean square of successive RR interval differences
- `hrv_coverage`: Percentage of valid measurements
- `hrv_low_frequency`: Low frequency HRV component
- `hrv_high_frequency`: High frequency HRV component
